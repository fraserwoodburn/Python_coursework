```python
import pandas as pd # import the pandas module so it can be used later

from sklearn.datasets import load_breast_cancer  
# load in the breast cancer dataset from the scikit learn sample datasets

cancer, target = load_breast_cancer(as_frame=True, return_X_y=True) # format the dataset
```

```python
cancer.head() # visualise the first five rows of the datset 
```

```python
cancer.plot()
```

```python
corr = cancer.corr() # make a correlation matrix of all the attributes in the dataset
corr # visualise the correlation matrix
```

```python
import seaborn as sns # import the seaborn package to enable better plotting

cmap = sns.diverging_palette(10,220, as_cmap=True)
sns.heatmap(corr, vmin=-1.0, vmax=1.0, cmap=cmap) # create a heatmap of correlating attributes
```

```python
from pandas.plotting import scatter_matrix
a = scatter_matrix(cancer, figsize=(21, 21), c=target) 
# create a more detailed scatter matrix to determine correlating factors and give an overview of the attributes 
# most distinctly defined between groups
```
