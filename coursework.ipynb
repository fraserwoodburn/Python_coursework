```python
import seaborn as sns # import the seaborn package to enable better plotting

cmap = sns.diverging_palette(10,220, as_cmap=True)
sns.heatmap(corr, vmin=-1.0, vmax=1.0, cmap=cmap) # create a heatmap of correlating attributes
```

```python
from pandas.plotting import scatter_matrix
a = scatter_matrix(cancer, figsize=(25, 25), c=target) 
# create a more detailed scatter matrix to determine correlating factors and give an overview of the attributes 
# most distinctly defined between groups
```

Now begin to investigate some of the attributes that in the scatter plot above look to be reasonably distinctly grouped between the two classes, benign and malignant cancers

```python
X = pd.DataFrame(cancer, columns=["worst concave points", "worst radius"]) # make a dataframe of two attributes
y = target

X.plot.scatter("worst concave points", "worst radius", c=target, colormap="plasma", colorbar=False) # plot the relationship - looks reasonable
```

```python
from sklearn.model_selection import train_test_split # import the train_test_split utility 

train_X, test_X, train_y, test_y = train_test_split(X, y, random_state=8) # split data into training dataset and test dataset, using random state of 8 for all going forward, just as means of consistency
```

```python
from sklearn import preprocessing # import preprocessing utility 
import numpy as np # import the numpy module
X_train = np.array([[ 1., -1.,  2.],
                     [ 2.,  0.,  0.],
                     [ 0.,  1., -1.]])
scaler = preprocessing.StandardScaler().fit(X_train)
scaler
scaler.mean_
scaler.scale_
X_scaled = scaler.transform(X_train)
X_scaled
X_scaled.mean(axis=0)
X_scaled.std(axis=0) # begin formatting the nearest neighbours classification to account for the difference in axis scale between the two attribuites chosen for the classification
```

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler # import utilities required to continue the scaling pipeline

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=8) # keep the random state as 8 for consistency
pipe = make_pipeline(StandardScaler(), LogisticRegression())
pipe.fit(X_train, y_train)  

pipe.score(X_test, y_test) # what is the new model score after scaling - looks very good, this could be the best two attributes to use for classification, but we'll explore a couple more just to confirm
```

```python
def plot_knn(model, X, y, resolution=100, ax=None):
    import matplotlib.pyplot as plt # import the matplotlib module

    x1 = X.iloc[:,0]
    x2 = X.iloc[:,1]
    x1_range = np.linspace(x1.min()*1.1 - x1.max()*0.1, x1.max()*1.1 - x1.min()*0.1, resolution)
    x2_range = np.linspace(x2.min()*1.1 - x2.max()*0.1, x2.max()*1.1 - x2.min()*0.1, resolution)
    grid_x1_values, grid_x2_values = np.meshgrid(x1_range, x2_range)
    x_prime = np.column_stack((grid_x1_values.ravel(), grid_x2_values.ravel()))
    x_prime = pd.DataFrame(x_prime, columns=X.columns)
    y_hat = model.predict(x_prime).reshape(grid_x1_values.shape)

    if ax is None:
        fig, ax = plt.subplots()
    ax.pcolormesh(grid_x1_values, grid_x2_values, y_hat, cmap="Pastel2", alpha=1.0, shading="auto")
    X.plot.scatter(0, 1, c=y, colormap="Dark2", colorbar=False, alpha=0.8, ax=ax)

plot_knn(pipe, X, y) # plot of the scaled model produced above, with each colour denoting the nearest neighbours classification for the group
```
